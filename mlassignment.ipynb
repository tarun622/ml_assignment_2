```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
plt.style.use('fivethirtyeight')
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
dataset=pd.read_csv('titanic.csv')
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
sb.countplot('Survived',data=dataset)
plt.show()
```

```python
dataset.groupby(['Sex', 'Survived'])['Survived'].count()
```

```python
sb.countplot('Pclass', hue='Survived', data=dataset)
plt.title('Pclass: Sruvived vs Dead')
plt.show()
```

```python
print('Oldest person Survived was of:',dataset['Age'].max())
print('Youngest person Survived was of:',dataset['Age'].min())
print('Average person Survived was of:',dataset['Age'].mean())
```
